#2.HAFTA: BULANIK MANTIK

***Kesin olmayan belirsiz seçenekleri, belirli kurallar çerçevesinde derecelendirme yeteneğine sahip olan bir mantık sistemidir.***

Bulanık mantık kavramı klasik yanlış (0) ve doğru (1) mantığı arasında derecelendirme yaparak kümelendirme işlemini gerçekleştirir. Problemin çözümü esnasında bazı önermelerin doğru ya da yanlış olarak değerlendirilmesi mümkün olmayabilir.

Bu durumun sebebi, önermelerin doğru bir şekilde ölçülememesinden kaynaklanmaktadır. İnsan beyninde gerçek durumlar daha çok ara değerler ile temsil edilebileceği için klasik (keskin) mantığa karşı “bulanık mantık” kavramı öne sürülmektedir.


**Bulanık Mantığın Aşamaları:**
* **Giriş**: Yapay zekâ bulanık mantık sistemlerinde ilk olarak veri girişi yapılır. Böylece bulanık mantık sistemi çalışmaya başlar.
* **Bulanıklaştırıcı**: Bulanık mantık sistemine girilen verileri 0 ile 1 arasındaki bulanık değerlere
dönüştürür.
* **Kurallar**: Bulanıklaştırma işlemi sırasındaki çıkarım kuralları belirlenir.
* **Çıkarım**: Bulanık kuralları kullanarak bulanık giriş değerleri için bulanık sonuçlar çıkarır.
* **Durulaştırıcı**: Çıkarımdan elde edilen sonuçları istenilen gerçek değerlere dönüştürür.
* **Çıkış**: Bulanık mantık sisteminden elde edilen çıkış verisidir.






## 1.Bulanık Mantık ile Trafik Sinyalizasyonu Örneği


* Kavşaktaki yaya yoğunluğuna bağlı olarak trafik lambaların sürelerini bulanık mantık
ile modeller.

* Yaya trafik sinyalizasyonuna ait bulanık model ile yaya sayısına göre en uygun yeşil ışık yanma
süresi belirlenmesi amaçlanmıştır.
* Öğrenciler yaya trafik sinyalizasyonuna ait giriş-çıkış parametrelerine göre tüm parametrelerin
üyelik fonksiyon sayılarını, isimlerini, alt ve üst limitlerini belirleyerek Python kodlarını hazırlar.

### 1.Bulanık Mantık ile Trafik Sinyalizasyonu Kodu


1. **`Antecedent` ve `Consequent` nesneleri**, sırasıyla giriş ve çıkış değişkenlerini temsil eder.
2. **`automf` fonksiyonu**, üyelik fonksiyonlarını otomatik olarak atar; bu örnekte her değişken için üçer adet üyelik fonksiyonu tanımlanmıştır: 'poor', 'average', 'good'.
3. **`Kural seti`,** eğer-ise mantığına dayalı olarak belirtilen koşullara göre çıkış değerlerini belirler.
4. **`ControlSystem` ve `ControlSystemSimulation` nesneleri**, kural setini içerir ve simülasyon için gereklidir.
5. **`input` metodu**, giriş değerlerini belirler.
6. **`compute` metodu**, belirtilen giriş değerlerine göre çıkış değerlerini hesaplar.
7. **`output` metodu**, çıkış değerlerine erişimi sağlar.
8. **`view` fonksiyonu**, sonuçları grafik olarak görselleştirmek için kullanılır.

````python
# Önce gerekli kütüphaneler import ediliyor.
import numpy as mat
import skfuzzy as bulanik
from skfuzzy import control as kontrol

# Yaya sayısı ve geçiş süresi için bulanık mantık değişkenleri oluşturuluyor.
yaya_sayisi = kontrol.Antecedent(mat.arange(1, 25, 1), 'yaya sayısı')
bekleme_suresi = kontrol.Consequent(mat.arange(15, 90, 1), 'geçiş zamanı')

# Yaya sayısı ve bekleme süresi için üçer üyelik fonksiyonu atanıyor (poor, average, good).
yaya_sayisi.automf(3)
bekleme_suresi.automf(3)

# Kural tabanı oluşturuluyor. Yani, "eğer yaya sayısı 'poor' ise, bekleme süresi de 'poor' olmalı" gibi kural seti oluşturuluyor.
kural1 = kontrol.Rule(yaya_sayisi['poor'], bekleme_suresi['poor'])
kural2 = kontrol.Rule(yaya_sayisi['average'], bekleme_suresi['average'])
kural3 = kontrol.Rule(yaya_sayisi['good'], bekleme_suresi['good'])

# Kontrol sistemi oluşturuluyor ve kural seti bu kontrol sistemine ekleniyor.
kontrol_sure = kontrol.ControlSystem([kural1, kural2, kural3])

# Kontrol sistemini simüle edecek bir nesne oluşturuluyor.
sure = kontrol.ControlSystemSimulation(kontrol_sure)

# Giriş değerini belirler (örneğin, yaya sayısı = 10)
sure.input['yaya sayısı'] = 1

# Simülasyonu çalıştırır
sure.compute()

# Elde edilen geçiş zamanı çıkış değerini yazdırır
print("Geçiş Zamanı:", sure.output['geçiş zamanı'])

# Grafik üzerinde sonuçları görselleştirmek için kullanılır.
bekleme_suresi.view(sim=sure)
````

````python
import numpy as mat
import skfuzzy as bulanik
from skfuzzy import control as kontrol

#giris
bulasik_miktari = kontrol.Antecedent(mat.arange(0, 100, 1), 'bulaşık miktarı')
kirlilik = kontrol.Antecedent( mat.arange(0, 100, 1), 'kirlilik seviyesi')

#cikis
yikama = kontrol.Consequent(mat.arange(0, 180, 1),'yıkama süresi')

bulasik_miktari['az'] = bulanik.trimf(bulasik_miktari.universe, [0, 0, 30])
bulasik_miktari['normal'] = bulanik.trimf(bulasik_miktari.universe, [10, 30, 60])
bulasik_miktari['çok'] = bulanik.trimf(bulasik_miktari.universe, [50, 60, 100])
kirlilik['az'] = bulanik.trimf(kirlilik.universe, [0, 0, 30])
kirlilik['normal'] = bulanik.trimf(kirlilik.universe, [10, 30, 60])
kirlilik['çok'] = bulanik.trimf(kirlilik.universe, [50, 60, 100])
yikama['kisa'] = bulanik.trimf(yikama.universe, [0, 0, 50])
yikama['normal'] = bulanik.trimf(yikama.universe, [40, 50, 100])
yikama['uzun'] = bulanik.trimf(yikama.universe, [60, 80, 180])

kural1 = kontrol.Rule(bulasik_miktari['az'] & kirlilik['az'], yikama['kisa'])
kural2 = kontrol.Rule(bulasik_miktari['normal'] & kirlilik['az'], yikama['normal'])
kural3 = kontrol.Rule(bulasik_miktari['çok'] & kirlilik['az'], yikama['normal'])
kural4 = kontrol.Rule(bulasik_miktari['az'] & kirlilik['normal'], yikama['normal'])
kural5 = kontrol.Rule(bulasik_miktari['normal'] & kirlilik['normal'], yikama['uzun'])
kural6 = kontrol.Rule(bulasik_miktari['çok'] & kirlilik['normal'], yikama['uzun'])
kural7 = kontrol.Rule(bulasik_miktari['az'] & kirlilik['çok'], yikama['normal'])
kural8 = kontrol.Rule(bulasik_miktari['normal'] & kirlilik['çok'], yikama['uzun'])
kural9 = kontrol.Rule(bulasik_miktari['çok'] & kirlilik['çok'], yikama['uzun'])

sonuc = kontrol.ControlSystem([kural1, kural2, kural3, kural4, kural5, kural6, kural7, kural8,
kural9])

model_sonuc = kontrol.ControlSystemSimulation(sonuc)


model_sonuc.input['bulaşık miktarı'] = 100
model_sonuc.input['kirlilik seviyesi']= 100

model_sonuc.compute()

print (model_sonuc.output['yıkama süresi'])
````

**Antecedent (Öncül)**: Bir mantıksal ifadenin başlangıç koşulu ya da girdisi olarak kullanılan şart veya durum. Bulanık mantıkta, antecedent genellikle bir kuralın sol tarafındaki koşulları ifade eder.

### Bulanık Mantık Üyelik Fonksiyonları ?

Üyelik fonksiyonları, bir değişkenin belirli bir kümenin içinde ne kadar olduğunu gösteren bir grafik gibidir. Üçgen ve dörtgen şeklinde iki tür üyelik fonksiyonu var. Üçgen, üç kenarlı bir şekle benzerken dörtgen, dört kenarlı bir şekle benzer.

Mesela, arabamızın hızını düşünelim. "Çok yavaş", "yavaş", "hızlı" ve "çok hızlı" gibi durumları değerlendirirken, hız değerlerini bir kümeye yerleştirmemiz gerekiyor. Mesela "hızlı" dediğimiz zaman ne kadar hızlıyı kast ediyoruz? İşte üyelik fonksiyonları bu konuda yardımcı olur.

u.png

**Üçgen Üyelik Fonksiyonu (trimf):**
Üçgen üyelik fonksiyonunda, üç ara değer belirleriz: başlangıç, orta ve bitiş değerleri. Mesela, "çok yakın" bir kümeyi ele alalım. Eğer araba, bir engelden 10 metre veya daha az bir mesafede ise, bu "çok yakın" kümeye aittir. Biz üyelik fonksiyonunu, arabanın o kümeye ne kadar uygun olduğunu gösterirken, 0'dan başlayarak 10'a kadar olan değerlerin bu kümeye ait olma derecesini belirliyoruz. Bu şekilde, mesafenin hangi değerlerde o kümenin içinde olduğunu daha net bir şekilde belirlemiş oluyoruz.

**Trapezoidal Üyelik Fonksiyonu (trapmf):**
Trapezoidal üyelik fonksiyonunda ise, başlangıç, orta iki değer ve bitiş değeri belirleriz. Örneğin, "çok yavaş" bir hız kategorisini ele alalım. Eğer araba, 0'dan başlayarak 30'a kadar olan bir hızda ise, bu "çok yavaş" kümeye aittir. Biz üyelik fonksiyonunu, arabanın o kümeye ne kadar uygun olduğunu gösterirken, 0'dan başlayarak 30'a kadar olan değerlerin bu kümeye ait olma derecesini belirliyoruz. Bu şekilde, hangi hız değerlerinin o kümeye daha uygun olduğunu belirlemiş oluyoruz.

````python
# Bu kod, bir araçtaki mesafe ve hız değerlerine dayalı olarak fren basıncını simüle eder.

# NumPy ve scikit-fuzzy kütüphaneleri içe aktarılır.
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl

# Antecedent'ler ve Consequent oluşturulur: Mesafe, Hız ve Fren Basıncı.
mesafe = ctrl.Antecedent(np.arange(0, 50, 1), 'mesafe')                    # Mesafe için 0 ile 50 arasında 1 birimlik artışla değerler oluşturulur.
hiz = ctrl.Antecedent(np.arange(0, 100, 1), 'hiz')                         # Hız için 0 ile 100 arasında 1 birimlik artışla değerler oluşturulur.
fren_basinci = ctrl.Consequent(np.arange(0, 100, 1), 'fren_basinci')       # Fren basıncı için 0 ile 100 arasında 1 birimlik artışla değerler oluşturulur.
````

````python

#ÜYELİK FONKSIYONLARI OLUŞTURULUR
mesafe['çok yakın'] = fuzz.trimf(mesafe.universe, [0, 5, 10])
mesafe['yakın'] = fuzz.trimf(mesafe.universe, [5, 15, 25])
mesafe['uzak'] = fuzz.trimf(mesafe.universe, [20, 30, 40])
mesafe['çok uzak'] = fuzz.trimf(mesafe.universe, [35, 45, 50])

# Hız için bulanık küme terimleri atanır.
hiz['çok yavaş'] = fuzz.trapmf(hiz.universe, [0, 0, 20, 30])
hiz['yavaş'] = fuzz.trapmf(hiz.universe, [20, 30, 45, 55])
hiz['hızlı'] = fuzz.trapmf(hiz.universe, [45, 55, 70, 80])
hiz['çok hızlı'] = fuzz.trapmf(hiz.universe, [70, 80, 90, 100])

# Fren basıncı için bulanık küme terimleri atanır.
fren_basinci['çok düşük'] = fuzz.trimf(fren_basinci.universe, [0, 10, 20])
fren_basinci['düşük'] = fuzz.trimf(fren_basinci.universe, [20, 30, 50])
fren_basinci['yüksek'] = fuzz.trimf(fren_basinci.universe, [45, 60, 70])
fren_basinci['çok yüksek'] = fuzz.trimf(fren_basinci.universe, [90, 95, 100])
````

````python
# Kurallar oluşturulur.
kural1 = ctrl.Rule(mesafe['çok yakın'] & hiz['çok yavaş'], fren_basinci['çok yüksek'])
kural2 = kontrol.Rule(mesafe['yakın'] & hiz['çok yavaş'] , fren_basinci['çok düşük'])
kural3 = kontrol.Rule(mesafe['çok yakın'] & hiz['yavaş'] , fren_basinci['çok yüksek'])
kural4 = kontrol.Rule(mesafe['yakın'] & hiz['yavaş'] , fren_basinci['düşük'])
kural5 = kontrol.Rule(mesafe['uzak'] & hiz['yavaş'] , fren_basinci['çok düşük'])
kural6 = kontrol.Rule(mesafe['çok yakın'] & hiz['hızlı'] , fren_basinci['çok yüksek'])
kural7 = kontrol.Rule(mesafe['yakın'] & hiz['hızlı'] , fren_basinci['düşük'])
kural8 = kontrol.Rule(mesafe['uzak'] & hiz['hızlı'] , fren_basinci['çok düşük'])
kural9 = kontrol.Rule(mesafe['çok yakın'] & hiz['çok hızlı'] , fren_basinci['çok yüksek'])
kural10 = kontrol.Rule(mesafe['yakın'] & hiz['çok hızlı'] , fren_basinci['yüksek'])
kural11 = kontrol.Rule(mesafe['uzak'] & hiz['çok hızlı'] , fren_basinci['düşük'])
kural12 = kontrol.Rule(mesafe['çok uzak'] & hiz['çok hızlı'] , fren_basinci['çok düşük'])

# Kontrol sistemine kurallar eklenir.
fren_kontrol = ctrl.ControlSystem([kural1, kural2, kural3, kural4, kural5, kural6, kural7, kural8, kural9, kural10, kural11, kural12])
````

````python

# Simülasyon oluşturulur.
frenleme = ctrl.ControlSystemSimulation(fren_kontrol)

````

````python

# Kullanıcıdan hız ve mesafe girişi alınır.
v = int(input("Hızı gir (0-100 km/h): "))
s = int(input("Mesafeyi gir (m): "))

# Basit bir koşul ile fren basılmasına gerek olup olmadığı kontrol edilir.
if (v/2 <= s):
    print("Fren basılması gerek yoktur.")
else:
    # Fren simülasyonu için gerekli girişler atanır.
    frenleme.input['hiz'] = v
    frenleme.input['mesafe'] = s

    # Fren basıncı hesaplanır ve sonuçlar yazdırılır.
    frenleme.compute()
    print("Fren basıncı (%): ", frenleme.output['fren_basinci'])
    print("Yeni hız değeri: ", v - (v * frenleme.output['fren_basinci'] / 100))
````

# 4.HAFTA: Karar Ağaçları

Karar ağaçları, veri madenciliği ve makine öğrenmesinde yaygın olarak kullanılan bir sınıflandırma ve regresyon algoritmasıdır.


1. **Karar Ağacı Nedir?**
   - Karar ağaçları, veri setindeki özelliklere (features) dayanarak bir dizi karar kuralı oluşturarak veri örneklerini sınıflandıran veya regresyon yapan bir ağaç yapısıdır.

2. **Karar Ağacı Yapısı:**
   - Her iç düğüm (node) bir karar noktasını temsil eder. Her yaprak düğüm (leaf node) ise bir sınıf etiketi veya regresyon tahmini içerir.
  
3. **Bölme Kriterleri:**
   - Karar ağaçları, veri setini bölme kriterlerine göre ağacın dallarına ayırır. Bölme kriterleri, veri setini homojen alt gruplara ayırmak için kullanılır. Gini impurity, entropi ve hata oranı gibi kriterler kullanılabilir.

4. **Karar Ağacı Oluşturma:**
   - Ağaç, veri setindeki özellikleri ve hedef değişkeni kullanarak en iyi bölme kriterini belirleyerek adım adım oluşturulur. Bu işlem, her bir düğümde en iyi bölme kriterini seçen ve ağacı oluşturan bir rekürsif süreçtir.

5. **Ağaç Pruning (Dalların Budanması):**
   - Ağaç, veri setine aşırı uyum (overfitting) eğiliminde olabilir. Bu durumu engellemek için ağaç, gereksiz dallardan kurtulmak için pruning işlemine tabi tutulabilir.

6. **Avantajlar:**
   - İnsanlar tarafından anlaşılabilir bir yapıya sahiptir.
   - Ağaç yapılarının yorumlanması ve görselleştirilmesi kolaydır.
   - Hem metinsel hem de sayısal veriler analiz edilebilir.
   - Veri ön işleme ihtiyacını azaltır.

7. **Dezavantajlar:**
   - Verilerin analizi yapılırken otomatik oluşan büyük ağaç yapılarının aşırı karmaşık olmasından dolayı ağaç dallarının takibi zordur.
   - Aşırı öğrenme (over fitting) yaşanabilir.


8. **Uygulama Alanları:**
   - Sınıflandırma ve regresyon problemleri için kullanılır.
   - Karar destek sistemleri, tıp, finans ve pazarlama gibi birçok alanında yaygın olarak kullanılır.

ETYA, çeşitli alanlarda, örneğin, robotik, oyun geliştirme, ekonomi, sosyal bilimler ve biyoloji gibi alanlarda kullanılır.

Çevre, etmenin etkileşimde bulunduğu ve eylemlerinin sonuçlarını etkileyen her şeydir. ETYA, etmenlerin çevreleriyle etkileşimlerini modellemek ve analiz etmek için çeşitli teknikler kullanır.   - ETYA'nın temel hedeflerinden biri, etmenlerin çevrelerinde nasıl karar alacağını ve eylem gerçekleştireceğini öğrenmesini sağlamaktır. Bu, **reinforcement** learning gibi algoritmalar kullanılarak yapılabilir.
  
  - **Reinforcement learning**, etmenin eylemlerinin sonuçlarına göre öğrenmesini sağlayan bir algoritmadır. Etmen, bir eylem gerçekleştirdiğinde, aldığı **ödül veya cezaya** göre öğrenir. - **Otomasyon:** Endüstriyel süreçlerdeki etmenleri modelleyerek otomasyon sistemlerinin verimliliğini artırabilir.
- **Simülasyon:** Etmen tabanlı modeller, gerçek dünyadaki durumları simüle etmek için kullanılabilir, bu da eğitim veya planlama amaçlarına hizmet edebilir.

## 1. Q - Frozen Lake





````python
import gym
import numpy as np

# FrozenLake ortamını oluştur
ortam = gym.make('FrozenLake-v1', is_slippery=False)

# Q tablosunu oluştur, başlangıçta tüm değerler sıfırdır
Q = np.zeros([ortam.observation_space.n, ortam.action_space.n])
````

````python
Q
````

````python

# Öğrenme oranı (eta), indirim faktörü (gma) ve döngü sayısını belirle
eta = -10
gma = 100
dongu = 250

# Toplam ödül değerlerini saklamak için bir liste oluştur
odul_list = []

# Belirlenen döngü sayısı kadar eğitim döngüsü başlat
for i in range(dongu):
    # Ortamı sıfırla (başlangıç durumuna getir)
    s = ortam.reset()
    odul_tumu = 0
    d = False
    j = 0

    # Bir bölüm (episode) boyunca döngü
    while j < 99:
        # Ortamı görüntüle
        ortam.render()
        j += 1

        # Epsilon-greedy politikası kullanarak bir eylem seç
        a = np.argmax(Q[s, :] + np.random.randn(1, ortam.action_space.n) * (1.0 / (i + 1)))

        # Seçilen eylemi uygula ve yeni durumu, ödülü ve terminal durumu al
        s1, o, d, _ = ortam.step(a)

        # Q değerini güncelle: Q(s, a) = Q(s, a) + eta * (o + gma * max(Q(s', :)) - Q(s, a))
        Q[s, a] = Q[s, a] + eta * (o + gma * np.max(Q[s1, :]) - Q[s, a])

        # Toplam ödülü güncelle
        odul_tumu += o

        # Yeni durumu güncelle
        s = s1

        # Terminal duruma ulaşıldıysa döngüden çık
        if d == True:
            break

    # Bölüm sonunda toplam ödülü listeye ekle
    odul_list.append(odul_tumu)

    # Ortamı bir kez daha görüntüle
    ortam.render()

# Eğitim sonrasında elde edilen Q tablosunu ekrana yazdır
print("Elde edilen Q Tablosu:")
print(Q)

# Ortalama ödül değerini ekrana yazdır
print("Ortalama ödül değeri:" + str(sum(odul_list) / dongu))
````

````python
# Q-Learning Algoritması için Ortam ve Parametrelerin Tanımlanması
import numpy as np

ortam_satir_sayisi = 11
ortam_sutun_sayisi = 11
````

````python
# Engel Kontrolü,                           yazılan engel_mi fonksiyonu robotun hareket etmesi sonrası gelinen hücrenin engel (yani kargo indirilen siyah ya da hedef yeşil kare) olup olmadığını tespit etmektedir

def engel_mi(gecerli_satir_indeks, gecerli_sutun_indeks):
  if oduller[gecerli_satir_indeks, gecerli_sutun_indeks] == -1.:
    return False
  else:
    return True
````

````python
# Başlangıç Noktasının Belirlenmesi,        yazılan baslangic_belirle fonksiyonu robotun her yeni öğrenme döngüsünde rastgele beyaz karelerden birinden başlamasını sağlanmaktadır

def baslangic_belirle():
  gecerli_satir_indeks = np.random.randint(ortam_satir_sayisi)
  gecerli_sutun_indeks = np.random.randint(ortam_sutun_sayisi)
  while engel_mi(gecerli_satir_indeks, gecerli_sutun_indeks):
    gecerli_satir_indeks = np.random.randint(ortam_satir_sayisi)
    gecerli_sutun_indeks = np.random.randint(ortam_sutun_sayisi)
  return gecerli_satir_indeks, gecerli_sutun_indeks
````

````python
# Hedefe Ulaşmak İçin Sonraki Hareketin Belirlenmesi

'''yazılan sonraki_hareket_belirle fonksiyonu robotun mevcut konumundan sonra
hareket edebileceği sonraki konumu eğitim sonunda elde edilen Q tablosundan belirlemektedir.
--> Bu noktada buradaki kod yapısına özel olarak belirlenen epsilon değeri de şans faktörü olarak
dikkate alınmaktadır. Buna göre, epsilon değerinden küçük olan rastgele bir değer üretilirse puan
tablosundan maksimum değerin, eşit ya da büyük rastgele bir değer üretilirse de bir başka
rastgele değerin (puan tablosu dikkate alınmadan) döndürülmesi sağlanmaktadır.'''

def sonraki_hareket_belirle(gecerli_satir_indeks, gecerli_sutun_indeks, epsilon):
  if np.random.random() < epsilon:
    return np.argmax(q_degerleri[gecerli_satir_indeks, gecerli_sutun_indeks])
  else:
    return np.random.randint(4)
````

````python
# Sonraki Noktaya Gitme

'''yazılan sonraki_noktaya_git fonksiyonu sayesinde, robotun bir sonraki hareketi
satır sütun ekseninde belirlenmekte; yeni hareket noktasına göre robotun yukarı hareketinde satır
değeri azaltılıp, aşağı hareketinde artırılmakta, yine sağa harekette sütun değeri artırılıp, sola
harekette azaltılmaktadır.'''

def sonraki_noktaya_git(gecerli_satir_indeks, gecerli_sutun_indeks, hareket_indeks):
  yeni_satir_indeks = gecerli_satir_indeks
  yeni_sutun_indeks = gecerli_sutun_indeks
  if hareketler[hareket_indeks] == 'yukari' and gecerli_satir_indeks > 0:
    yeni_satir_indeks -= 1
  elif hareketler[hareket_indeks] == 'sag' and gecerli_sutun_indeks < ortam_sutun_sayisi - 1:
    yeni_sutun_indeks += 1
  elif hareketler[hareket_indeks] == 'asagi' and gecerli_satir_indeks < ortam_satir_sayisi - 1:
    yeni_satir_indeks += 1
  elif hareketler[hareket_indeks] == 'sol' and gecerli_sutun_indeks > 0:
    yeni_sutun_indeks -= 1
  return yeni_satir_indeks, yeni_sutun_indeks
````

````python
# En Kısa Mesafeyi Bulma,       Robotun Q-Öğrenme ile eğitimi sonrasında belirlenmiş olan en uygun yolu verecek fonksiyon yazılır

def en_kisa_mesafe(basla_satir_indeks, basla_sutun_indeks):
  if engel_mi(basla_satir_indeks, basla_sutun_indeks):
    return []
  else:
    gecerli_satir_indeks, gecerli_sutun_indeks = basla_satir_indeks, basla_sutun_indeks
    en_kisa = []
    en_kisa.append([gecerli_satir_indeks, gecerli_sutun_indeks])
    while not engel_mi(gecerli_satir_indeks, gecerli_sutun_indeks):
      hareket_indeks = sonraki_hareket_belirle(gecerli_satir_indeks, gecerli_sutun_indeks, 1.)
      gecerli_satir_indeks, gecerli_sutun_indeks = sonraki_noktaya_git(gecerli_satir_indeks,
                                                                       gecerli_sutun_indeks, hareket_indeks)
      en_kisa.append([gecerli_satir_indeks, gecerli_sutun_indeks])
    return en_kisa
````

````python
# Q-Learning Algoritması
epsilon = 0.9
azalma_degeri = 0.9
ogrenme_orani = 0.9

for adim in range(1000):
  satir_indeks, sutun_indeks = baslangic_belirle()
  while not engel_mi(satir_indeks, sutun_indeks):
    hareket_indeks = sonraki_hareket_belirle(satir_indeks, sutun_indeks, epsilon)
    eski_satir_indeks, eski_sutun_indeks = satir_indeks, sutun_indeks
    satir_indeks, sutun_indeks = sonraki_noktaya_git(satir_indeks, sutun_indeks, hareket_indeks)
    odul = oduller[satir_indeks, sutun_indeks]
    eski_q_degeri = q_degerleri[eski_satir_indeks, eski_sutun_indeks, hareket_indeks]
    fark = odul + (azalma_degeri * np.max(q_degerleri[satir_indeks, sutun_indeks])) - eski_q_degeri
    yeni_q_degeri = eski_q_degeri + (ogrenme_orani * fark)
    q_degerleri[eski_satir_indeks, eski_sutun_indeks, hareket_indeks] = yeni_q_degeri

print('Eğitim tamamlandı.')
````

# 7.Hafta Zeki Optimizasyon

**Optimizasyon nedir?**
- Belirli bir hedefe veya amaca ulaşmak için kaynakları en iyi şekilde kullanma süreci
- Gerçek dünya problemlerini çözmek ve verimliliği artırmak için sıklıkla kullanılır

**Optimizasyonun önemi**
- Çeşitli sektörlerde kaynakların daha verimli kullanılmasını sağlar
- Maliyetleri düşürür, üretkenliği artırır ve kaliteyi iyileştirir
- Karar verme süreçlerini daha efektif hale getirir  

**Zeki optimizasyon algoritmalarının bu zorlukları aşmadaki rolü**
- Geleneksel yöntemlerin yetersiz kaldığı durumlarda kullanılabilir
- Karmaşık ve çok boyutlu arama alanlarında etkilidir
- Genellikle iyi bir çözüme yakın başlar ve daha sonra bu çözümü iyileştirir
- Tutarlı ve güvenilir sonuçlar üretirler


**Bölüm 1 - Genetik Algoritmalar**


**Genetik algoritmaların temel prensipleri**
- Doğal seçilim ve evrimin mekanizmalarından esinlenir
- Bir çözüm kümesinden, belirli kriterlere göre seçilen çözümler üzerinde çaprazlama ve mutasyon gibi işlemler yapılarak yeni çözümler üretilir
- Bu süreç, belirli bir hedefe ulaşıncaya veya belirli kriterler karşılanana kadar tekrarlanır

**Bireyler, genler ve popülasyonlar kavramları**
- Genetik algoritmalarda her çözüm bir birey olarak temsil edilir

## 7.H Genetik Optimizasyon Denklem Hesaplama

Bu kod, PyGAD kütüphanesini kullanarak genetik algoritma ile bir denklemin en iyi çözümünü bulma işlemini gerçekleştirir. İşlev ve parametrelerin detaylı açıklamaları aşağıda bulunmaktadır:

**giris_degerleri**: Denklemin katsayılarını içeren bir liste. Örneğin, [4, -2, 3.5, 5, -11, -4.7] listesi, 4x - 2y + 3.5z + 5k - 11m - 4.7n = 44 denkleminin katsayılarını temsil eder.

**beklenen_cikis**: Denklemdeki bilinmeyenleri kullanarak hesaplanan ve beklenen çıkış değerini temsil eder. Bu örnekte 44 olarak belirlenmiştir.

**uygunluk_fonksiyonu**: Genetik algoritmanın her bir bireyin ne kadar iyi olduğunu ölçmek için kullanılan fonksiyon. Bu fonksiyon, bireyin denkleme verdiği çıkışın beklenen çıkışa ne kadar yakın olduğunu ölçer. Uygunluk değeri, bu farkın tersine alınmış hali olarak hesaplanır, çünkü genetik algoritma, uygunluk değeri büyüdükçe performansın artmasını bekler.

**nesil_sayisi**: Genetik algoritmanın kaç nesil boyunca çalışacağını belirten parametre.

**ebeveyn_secmesi_sayisi**: Her bir nesilde çocuk oluşturmak için seçilecek ebeveyn sayısını belirten parametre.

**populasyon_sayisi**: Her bir nesilde kaç bireyin bulunacağını belirten parametre.

**gen_sayisi**: Her bir bireyin gen sayısını (parametre sayısı) belirten parametre.

**son_uygunluk_degeri**: Her nesil sonunda çıktı olarak yazdırılan en iyi uygunluk değerini tutan bir değişken.

**nesil_ozeti_bilgisi**: Her nesil sonunda çağrılan bir fonksiyon. Bu fonksiyon, genetik algoritmanın ilerlemesini izlemek ve çıktıları görmek için kullanılır.

**genetik_algoritma**: PyGAD kütüphanesinin GA sınıfından bir nesne oluşturulur. Bu nesne, genetik algoritmanın parametrelerini ve diğer ayarlarını içerir.

**genetik_algoritma.run()**: Genetik algoritmayı çalıştıran metod. Algoritma burada nesil nesil ilerler ve en iyi çözümü bulmaya çalışır.

**genetik_algoritma.best_solution()**: En iyi çözümü, çözümün uygunluğunu ve çözümün indeksini döndüren metod.

**cozum**: En iyi çözümü temsil eden bir dizi. Bu dizinin elemanları, denklemin bilinmeyenlerine karşılık gelen değerleri içerir.

**tahmin**: En iyi çözümü kullanarak denklemi çözüp elde edilen tahmini çıkış değeri.

**genetik_algoritma.best_solution_generation**: En iyi çözümün hangi nesilde elde edildiğini belirten bir özellik. Eğer -1 ise, en iyi çözüm henüz bulunmamıştır.


Kodun çıktısı, her nesil sonunda nesil numarası, fonksiyon sonucu, ve değişim miktarını içerir. En sonunda, en iyi çözümün değişken değerleri, indeks numarası ve tahmini çıkış değeri yazdırılır. Eğer en iyi çözüm bulunduysa, kaçıncı nesilde bulunduğu da belirtilir.

````python
import pygad
import numpy

# Fonksiyon giriş değerleri ve beklenen çıkış değeri
giris_degerleri = [2,3]          # Denklem ( 4x - 2y + 3.5z + 5k - 11m - 4.7n = 44 ), (2x+3y = 12)
beklenen_cikis = 12

# Uygunluk fonksiyonu tanımı
def uygunluk_fonksiyonu(ga_instance, cozum, cozum_idx):
    cikis = numpy.sum(cozum * giris_degerleri)
    uygunluk = 1.0 / numpy.abs(cikis - beklenen_cikis)
    return uygunluk


# Genetik algoritmanın parametreleri
nesil_sayisi = 100
ebeveyn_secmesi_sayisi = 7

populasyon_sayisi = 10
gen_sayisi = len(giris_degerleri)

son_uygunluk_degeri = 0

# Her nesil sonunda çağrılacak fonksiyon
def nesil_ozeti_bilgisi(ga_instance):
    global son_uygunluk_degeri
    print("Nesil = ", ga_instance.generations_completed)
    print("Fonksiyon Sonucu = ", ga_instance.best_solution()[1])
    print("Değişim = ", ga_instance.best_solution()[1] - son_uygunluk_degeri)
    son_uygunluk_degeri = ga_instance.best_solution()[1]
````

````python

# Genetik algoritma örneği oluşturma
genetik_algoritma = pygad.GA(num_generations=nesil_sayisi,
                             num_parents_mating=ebeveyn_secmesi_sayisi,
                             fitness_func=uygunluk_fonksiyonu,
                             sol_per_pop=populasyon_sayisi,
                             num_genes=gen_sayisi,
                             on_generation=nesil_ozeti_bilgisi)

# Genetik algoritmayı çalıştırma
genetik_algoritma.run()

# En iyi çözümü alma
cozum, cozum_uygunluk, cozum_idx = genetik_algoritma.best_solution()

# Sonuçları yazdırma
print("En uygun çözüm değişken değerleri: ", cozum)
print("En uygun çözümü veren birey indeks no.:", cozum_idx)

tahmin = numpy.sum(numpy.array(giris_degerleri) * cozum)
print("En uygun çözüm ile fonksiyon sonucu: ", tahmin)

if genetik_algoritma.best_solution_generation != -1:
    print(f"En uygun çözüm {genetik_algoritma.best_solution_generation} nesil sonra elde edildi.")
````

## 7.H Karınca Kolonisi Mesafe Hesaplama

Karınca kolonisi algoritmasının performansını etkileyen birkaç önemli parametre vardır. Bu parametreler, algoritmanın ne kadar keşif ve ne kadar zahmetli çalışacağını belirler. İşte bazı önemli parametreler ve değerleri:

- **n_ants** (Karınca Sayısı): Kaç adet karınca kullanılacağını belirler. Genellikle küçük problemler için 5-10, orta büyüklükteki problemler için 10-20, büyük problemler için 20 ve üzeri karınca kullanmak uygun olabilir.

- **alpha ve beta** (Feromon ve Uzaklık Katsayıları): Feromon ve uzaklık katsayıları, karıncaların feromonları ve uzaklıkları nasıl değerlendireceklerini belirler. Genellikle alpha=1, beta=2 gibi bir değer kullanılır. Bu, feromonların uzaklıklardan daha fazla etkilenmesini sağlar.

- **pheromone_decay** (Feromonun Bozulma Oranı): Feromonun zamanla azalma oranını belirler. Genellikle 0.9 veya 0.95 gibi değerler kullanılır.

- **n_best** (En İyi Yol Sayısı): Her nesilde en iyi yolların sayısını belirler. Genellikle 1-5 arası bir değer kullanılabilir.

- **n_iterations** (İterasyon Sayısı): Algoritmanın kaç iterasyon boyunca çalışacağını belirler. Daha fazla iterasyon, daha fazla keşif anlamına gelir. Genellikle 100'den başlayarak artırılabilir.

- **decay** (Feromon Azalma Faktörü): Her iterasyonda feromonların azalma oranını belirler. Daha yüksek bir değer, feromonların daha hızlı azalmasına neden olur.

````python
import numpy

# Karınca Koloni Algoritması sınıfı
class AntColony(object):

    # __init__ fonksiyonu, sınıfın başlatılması için kullanılır.
    def __init__(self, distances, n_ants, n_best, n_iterations, decay, alpha=1, beta=1):
        # Uzaklık matrisi (AntColony'nin çalışma alanı) ve diğer parametrelerin atanması
        self.distances  = distances
        self.pheromone = np.ones(self.distances.shape) / len(distances)
        self.all_inds = range(len(distances))
        self.n_ants = n_ants
        self.n_best = n_best
        self.n_iterations = n_iterations
        self.decay = decay
        self.alpha = alpha
        self.beta = beta

    # Algoritmanın ana çalışma fonksiyonu
    def run(self):
        shortest_path = None
        all_time_shortest_path = ("placeholder", np.inf)
        # Belirlenen iterasyon sayısı kadar algoritmanın çalıştırılması
        for i in range(self.n_iterations):
            all_paths = self.gen_all_paths()
            # Feromon dağıtımı ve en kısa yolun güncellenmesi
            self.spread_pheronome(all_paths, self.n_best, shortest_path=shortest_path)
            shortest_path = min(all_paths, key=lambda x: x[1])
            print(shortest_path)
            if shortest_path[1] < all_time_shortest_path[1]:
                all_time_shortest_path = shortest_path
            self.pheromone * self.decay  # Feromonun azaltılması
        return all_time_shortest_path

    # Feromonun yayılması
    def spread_pheronome(self, all_paths, n_best, shortest_path):
        sorted_paths = sorted(all_paths, key=lambda x: x[1])
        # En iyi yolların feromonlarının güncellenmesi
        for path, dist in sorted_paths[:n_best]:
            for move in path:
                self.pheromone[move] += 1.0 / self.distances[move]

    # Belirli bir yolun uzunluğunun hesaplanması
    def gen_path_dist(self, path):
        total_dist = 0
        for ele in path:
            total_dist += self.distances[ele]
        return total_dist

    # Tüm karınca yollarının oluşturulması
    def gen_all_paths(self):
        all_paths = []
        # Belirli sayıda karınca için yol oluşturulması
        for i in range(self.n_ants):
            path = self.gen_path(0)
            all_paths.append((path, self.gen_path_dist(path)))
        return all_paths

    # Belirli bir karınca yolu oluşturulması
    def gen_path(self, start):
        path = []
        visited = set()
        visited.add(start)
        prev = start
        # Tüm şehirleri ziyaret edene kadar devam eden bir döngü
        for i in range(len(self.distances) - 1):
            move = self.pick_move(self.pheromone[prev], self.distances[prev], visited)
            path.append((prev, move))
            prev = move
            visited.add(move)
        path.append((prev, start))  # Geri dönüş eklenir
        return path

    # Bir sonraki hareketin seçilmesi
    def pick_move(self, pheromone, dist, visited):
        pheromone = np.copy(pheromone)
        pheromone[list(visited)] = 0

        row = pheromone ** self.alpha * ((1.0 / dist) ** self.beta)

        norm_row = row / row.sum()
        move = np_choice(self.all_inds, 1, p=norm_row)[0]
        return move
````
